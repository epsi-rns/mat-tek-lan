```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
import sympy as sp
import pandas as pd
import math

from IPython.display import display, Math# Set up the plotting style

try:
    plt.style.use('seaborn-whitegrid')
except:
    plt.style.use('default')  # Fallback to default style
```

# Basic

## Trigonometry Function

```python
# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Basic Trigonometric Functions', fontsize=16, fontweight='bold')

# Generate x values from -2π to 2π
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Plot sin(x)
axes[0].plot(x, np.sin(x), color='blue', linewidth=2)
axes[0].set_title('sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].set_ylim(-1.5, 1.5)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0].grid(True, alpha=0.3)

# Plot cos(x)
axes[1].plot(x, np.cos(x), color='teal', linewidth=2)
axes[1].set_title('cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].set_ylim(-1.5, 1.5)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)

# Plot tan(x) - handle asymptotes
tan_x = np.tan(x)
# Remove values that are too large for plotting
tan_x = np.where(np.abs(tan_x) > 10, np.nan, tan_x)
axes[2].plot(x, tan_x, color='red', linewidth=2)
axes[2].set_title('tan(x)')
axes[2].set_xlabel('x')
axes[2].set_ylabel('tan(x)')
axes[2].set_ylim(-5, 5)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[2].grid(True, alpha=0.3)

# Add vertical lines at multiples of π/2 for reference
for i in range(-4, 5):
    for ax in axes:
        ax.axvline(x=i*np.pi/2, color='gray', linestyle='--', alpha=0.5)

# Add π labels to x-axis
for ax in axes:
    ax.set_xticks([-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, 
                   np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels(['-2π', '-3π/2', '-π', '-π/2', '0', 
                        'π/2', 'π', '3π/2', '2π'])

plt.tight_layout()
plt.show()
```

## Reciprocal

```python
# Use matplotlib's built-in mathtext (doesn't require LaTeX)
plt.rcParams['mathtext.fontset'] = 'stix'  # Use STIX fonts for math

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Reciprocal Trigonometric Functions', fontsize=16, fontweight='bold')

# Generate x values from -2π to 2π, avoiding zeros in denominators
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
# Remove points very close to where sin(x) or cos(x) = 0 to avoid division by zero issues
x = x[(np.abs(np.sin(x)) > 0.01) & (np.abs(np.cos(x)) > 0.01)]

# Plot csc(x) = 1/sin(x)
csc_x = 1 / np.sin(x)
csc_x = np.where(np.abs(csc_x) > 10, np.nan, csc_x)  # Limit extreme values
axes[0].plot(x, csc_x, color='blue', linewidth=2)
axes[0].set_title(r'$\csc x = \frac{1}{\sin x}$', fontsize=14)
axes[0].set_xlabel(r'$x$')
axes[0].set_ylabel(r'$\csc x$')
axes[0].set_ylim(-5, 5)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot sec(x) = 1/cos(x)
sec_x = 1 / np.cos(x)
sec_x = np.where(np.abs(sec_x) > 10, np.nan, sec_x)  # Limit extreme values
axes[1].plot(x, sec_x, color='teal', linewidth=2)
axes[1].set_title(r'$\sec x = \frac{1}{\cos x}$', fontsize=14)
axes[1].set_xlabel(r'$x$')
axes[1].set_ylabel(r'$\sec x$')
axes[1].set_ylim(-5, 5)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot cot(x) = cos(x)/sin(x)
cot_x = np.cos(x) / np.sin(x)
cot_x = np.where(np.abs(cot_x) > 10, np.nan, cot_x)  # Limit extreme values
axes[2].plot(x, cot_x, color='red', linewidth=2)
axes[2].set_title(r'$\cot x = \frac{\cos x}{\sin x}$', fontsize=14)
axes[2].set_xlabel(r'$x$')
axes[2].set_ylabel(r'$\cot x$')
axes[2].set_ylim(-5, 5)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Add vertical lines at multiples of π/2 for reference
for i in range(-4, 5):
    for ax in axes:
        ax.axvline(x=i*np.pi/2, color='gray', linestyle='--', alpha=0.5)

# Add π labels to x-axis
for ax in axes:
    ax.set_xticks([-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, 
                   np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels([r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', 
                        r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])

plt.tight_layout()
plt.show()
```

## Inverse

```python
# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(r'Trigonometric Functions and Their Inverses', fontsize=16, fontweight='bold')

# Generate x values
x_trig = np.linspace(-2*np.pi, 2*np.pi, 1000)
x_inv = np.linspace(-1, 1, 500)  # For arcsin and arccos
x_atan = np.linspace(-10, 10, 1000)  # For arctan

# Row 1: Regular trigonometric functions
axes[0, 0].plot(x_trig, np.sin(x_trig), color='blue', linewidth=2)
axes[0, 0].set_title(r'$\sin x$', fontsize=14)
axes[0, 0].set_ylabel(r'Value')
axes[0, 0].set_ylim(-1.5, 1.5)
axes[0, 0].set_xlim(-2*np.pi, 2*np.pi)

axes[0, 1].plot(x_trig, np.cos(x_trig), color='teal', linewidth=2)
axes[0, 1].set_title(r'$\cos x$', fontsize=14)
axes[0, 1].set_ylim(-1.5, 1.5)
axes[0, 1].set_xlim(-2*np.pi, 2*np.pi)

axes[0, 2].plot(x_trig, np.tan(x_trig), color='red', linewidth=2)
axes[0, 2].set_title(r'$\tan x$', fontsize=14)
axes[0, 2].set_ylim(-5, 5)
axes[0, 2].set_xlim(-2*np.pi, 2*np.pi)

# Row 2: Inverse trigonometric functions
axes[1, 0].plot(x_inv, np.arcsin(x_inv), color='blue', linewidth=2)
axes[1, 0].set_title(r'$\arcsin x$', fontsize=14)
axes[1, 0].set_xlabel(r'$x$')
axes[1, 0].set_ylabel(r'Value')
axes[1, 0].set_xlim(-1.1, 1.1)
axes[1, 0].set_ylim(-np.pi/2 - 0.1, np.pi/2 + 0.1)

axes[1, 1].plot(x_inv, np.arccos(x_inv), color='teal', linewidth=2)
axes[1, 1].set_title(r'$\arccos x$', fontsize=14)
axes[1, 1].set_xlabel(r'$x$')
axes[1, 1].set_xlim(-1.1, 1.1)
axes[1, 1].set_ylim(-0.1, np.pi + 0.1)

axes[1, 2].plot(x_atan, np.arctan(x_atan), color='red', linewidth=2)
axes[1, 2].set_title(r'$\arctan x$', fontsize=14)
axes[1, 2].set_xlabel(r'$x$')
axes[1, 2].set_xlim(-10, 10)
axes[1, 2].set_ylim(-np.pi/2 - 0.1, np.pi/2 + 0.1)

# Add grid, zero lines, and format axes
for i in range(2):
    for j in range(3):
        axes[i, j].axhline(y=0, color='k', linestyle='-', alpha=0.3)
        axes[i, j].axvline(x=0, color='k', linestyle='-', alpha=0.3)
        axes[i, j].grid(True, alpha=0.3)
        
        # Add π labels to y-axis for inverse functions
        if i == 1:
            y_ticks = [-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]
            y_labels = [r'$-\frac{\pi}{2}$', r'$-\frac{\pi}{4}$', r'$0$', 
                        r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$']
            axes[i, j].set_yticks(y_ticks)
            axes[i, j].set_yticklabels(y_labels)
        
        # Add π labels to x-axis for trigonometric functions
        if i == 0:
            axes[i, j].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
            axes[i, j].set_xticklabels([r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$'])

plt.tight_layout()
plt.show()
```

```python
# Create a visualization showing the relationship between sin and arcsin
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle(r'Relationship Between $\sin x$ and $\arcsin x$', fontsize=16, fontweight='bold')

# Plot sin(x) and its inverse
x_sin = np.linspace(-np.pi/2, np.pi/2, 1000)
y_sin = np.sin(x_sin)

ax1.plot(x_sin, y_sin, color='blue', linewidth=3, label=r'$\sin x$')
ax1.set_title(r'$\sin x$ on restricted domain', fontsize=14)
ax1.set_xlabel(r'$x$ (angle)')
ax1.set_ylabel(r'$\sin x$')
ax1.set_xlim(-np.pi/2 - 0.1, np.pi/2 + 0.1)
ax1.set_ylim(-1.1, 1.1)
ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax1.grid(True, alpha=0.3)

# Add π labels to x-axis
ax1.set_xticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2])
ax1.set_xticklabels([r'$-\frac{\pi}{2}$', r'$-\frac{\pi}{4}$', r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$'])

# Plot the line y = x for reference
ax1.plot([-np.pi/2, np.pi/2], [-1, 1], color='red', linestyle='--', alpha=0.7, label=r'$y = x$ (if linear)')
ax1.legend()

# Second plot: Show both functions and their reflection property
x_asin = np.linspace(-1, 1, 1000)
y_asin = np.arcsin(x_asin)

ax2.plot(x_sin, y_sin, color='blue', linewidth=3, label=r'$\sin x$')
ax2.plot(y_asin, x_asin, color='green', linewidth=3, label=r'$\arcsin x$')
ax2.plot([-np.pi/2, np.pi/2], [-np.pi/2, np.pi/2], color='red', linestyle='--', alpha=0.7, label=r'Line $y = x$')

ax2.set_title(r'$\sin x$ and $\arcsin x$ are reflections across $y = x$', fontsize=14)
ax2.set_xlabel(r'$x$')
ax2.set_ylabel(r'$y$')
ax2.set_xlim(-np.pi/2 - 0.1, np.pi/2 + 0.1)
ax2.set_ylim(-1.1, 1.1)
ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)
ax2.grid(True, alpha=0.3)
ax2.legend()

# Add π labels to x-axis
ax2.set_xticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2])
ax2.set_xticklabels([r'$-\frac{\pi}{2}$', r'$-\frac{\pi}{4}$', r'$0$', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$'])

plt.tight_layout()
plt.show()
```

## Hyperbolic

```python
# Use matplotlib's built-in mathtext for LaTeX-like rendering
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'serif'

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(r'Hyperbolic Functions', fontsize=16, fontweight='bold')

# Generate x values from -3 to 3
x = np.linspace(-3, 3, 1000)

# Plot sinh(x) = (e^x - e^{-x})/2
sinh_x = (np.exp(x) - np.exp(-x)) / 2
axes[0].plot(x, sinh_x, color='blue', linewidth=2, label=r'$\sinh x$')
axes[0].plot(x, np.exp(x)/2, '--', color='gray', alpha=0.7, label=r'$\frac{1}{2}e^x$')
axes[0].plot(x, -np.exp(-x)/2, '--', color='gray', alpha=0.7, label=r'$-\frac{1}{2}e^{-x}$')
axes[0].set_title(r'$\sinh x = \frac{e^x - e^{-x}}{2}$', fontsize=14)
axes[0].set_xlabel(r'$x$')
axes[0].set_ylabel(r'$\sinh x$')
axes[0].set_ylim(-5, 5)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0].grid(True, alpha=0.3)
axes[0].legend()

# Plot cosh(x) = (e^x + e^{-x})/2
cosh_x = (np.exp(x) + np.exp(-x)) / 2
axes[1].plot(x, cosh_x, color='teal', linewidth=2, label=r'$\cosh x$')
axes[1].plot(x, np.exp(x)/2, '--', color='gray', alpha=0.7, label=r'$\frac{1}{2}e^x$')
axes[1].plot(x, np.exp(-x)/2, '--', color='gray', alpha=0.7, label=r'$\frac{1}{2}e^{-x}$')
axes[1].set_title(r'$\cosh x = \frac{e^x + e^{-x}}{2}$', fontsize=14)
axes[1].set_xlabel(r'$x$')
axes[1].set_ylabel(r'$\cosh x$')
axes[1].set_ylim(-1, 5)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)
axes[1].legend()

# Plot tanh(x) = (e^x - e^{-x})/(e^x + e^{-x})
tanh_x = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
axes[2].plot(x, tanh_x, color='red', linewidth=2, label=r'$\tanh x$')
axes[2].set_title(r'$\tanh x = \frac{e^x - e^{-x}}{e^x + e^{-x}}$', fontsize=14)
axes[2].set_xlabel(r'$x$')
axes[2].set_ylabel(r'$\tanh x$')
axes[2].set_ylim(-1.1, 1.1)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[2].grid(True, alpha=0.3)
axes[2].legend()

# Add horizontal asymptotes for tanh
axes[2].axhline(y=1, color='gray', linestyle='--', alpha=0.7)
axes[2].axhline(y=-1, color='gray', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle(r'Hyperbolic Functions and Their Exponential Components', fontsize=16, fontweight='bold')

# Generate x values
x = np.linspace(-3, 3, 1000)

# Top left: sinh(x) and its components
axes[0, 0].plot(x, (np.exp(x) - np.exp(-x)) / 2, color='blue', linewidth=3, label=r'$\sinh x = \frac{e^x - e^{-x}}{2}$')
axes[0, 0].plot(x, np.exp(x)/2, '--', color='orange', alpha=0.7, label=r'$\frac{1}{2}e^x$')
axes[0, 0].plot(x, -np.exp(-x)/2, '--', color='green', alpha=0.7, label=r'$-\frac{1}{2}e^{-x}$')
axes[0, 0].set_title(r'Hyperbolic Sine: $\sinh x$', fontsize=14)
axes[0, 0].set_xlabel(r'$x$')
axes[0, 0].set_ylabel(r'$y$')
axes[0, 0].set_ylim(-5, 5)
axes[0, 0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0, 0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].legend()

# Top right: cosh(x) and its components
axes[0, 1].plot(x, (np.exp(x) + np.exp(-x)) / 2, color='teal', linewidth=3, label=r'$\cosh x = \frac{e^x + e^{-x}}{2}$')
axes[0, 1].plot(x, np.exp(x)/2, '--', color='orange', alpha=0.7, label=r'$\frac{1}{2}e^x$')
axes[0, 1].plot(x, np.exp(-x)/2, '--', color='green', alpha=0.7, label=r'$\frac{1}{2}e^{-x}$')
axes[0, 1].set_title(r'Hyperbolic Cosine: $\cosh x$', fontsize=14)
axes[0, 1].set_xlabel(r'$x$')
axes[0, 1].set_ylabel(r'$y$')
axes[0, 1].set_ylim(-1, 5)
axes[0, 1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0, 1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].legend()

# Bottom left: tanh(x)
axes[1, 0].plot(x, (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)), color='red', linewidth=3, label=r'$\tanh x = \frac{e^x - e^{-x}}{e^x + e^{-x}}$')
axes[1, 0].set_title(r'Hyperbolic Tangent: $\tanh x$', fontsize=14)
axes[1, 0].set_xlabel(r'$x$')
axes[1, 0].set_ylabel(r'$\tanh x$')
axes[1, 0].set_ylim(-1.1, 1.1)
axes[1, 0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1, 0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].legend()

# Add horizontal asymptotes for tanh
axes[1, 0].axhline(y=1, color='gray', linestyle='--', alpha=0.7, label=r'$y = 1$')
axes[1, 0].axhline(y=-1, color='gray', linestyle='--', alpha=0.7, label=r'$y = -1$')
axes[1, 0].legend()

# Bottom right: All three hyperbolic functions together
axes[1, 1].plot(x, np.sinh(x), color='blue', linewidth=2, label=r'$\sinh x$')
axes[1, 1].plot(x, np.cosh(x), color='teal', linewidth=2, label=r'$\cosh x$')
axes[1, 1].plot(x, np.tanh(x), color='red', linewidth=2, label=r'$\tanh x$')
axes[1, 1].set_title(r'All Hyperbolic Functions', fontsize=14)
axes[1, 1].set_xlabel(r'$x$')
axes[1, 1].set_ylabel(r'$y$')
axes[1, 1].set_ylim(-2, 4)
axes[1, 1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1, 1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

## Hyperbolic Reciprocal

```python
# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(r'Hyperbolic Functions and Their Reciprocals', fontsize=16, fontweight='bold')

# Generate x values, avoiding zeros where functions are undefined
x = np.linspace(-3, 3, 1000)
x_pos = x[x > 0.01]  # Positive x values (for functions undefined at 0)
x_neg = x[x < -0.01]  # Negative x values (for functions undefined at 0)

# Row 1: Original hyperbolic functions
axes[0, 0].plot(x, np.sinh(x), color='blue', linewidth=2)
axes[0, 0].set_title(r'$\sinh x$', fontsize=14)
axes[0, 0].set_ylabel(r'Value')
axes[0, 0].set_ylim(-5, 5)

axes[0, 1].plot(x, np.cosh(x), color='teal', linewidth=2)
axes[0, 1].set_title(r'$\cosh x$', fontsize=14)
axes[0, 1].set_ylim(0, 5)

axes[0, 2].plot(x, np.tanh(x), color='red', linewidth=2)
axes[0, 2].set_title(r'$\tanh x$', fontsize=14)
axes[0, 2].set_ylim(-1.1, 1.1)

# Row 2: Reciprocal hyperbolic functions
axes[1, 0].plot(x_neg, 1/np.sinh(x_neg), color='blue', linewidth=2)
axes[1, 0].plot(x_pos, 1/np.sinh(x_pos), color='blue', linewidth=2)
axes[1, 0].set_title(r'$\operatorname{csch} x = \frac{1}{\sinh x}$', fontsize=14)
axes[1, 0].set_xlabel(r'$x$')
axes[1, 0].set_ylabel(r'Value')
axes[1, 0].set_ylim(-5, 5)

axes[1, 1].plot(x, 1/np.cosh(x), color='teal', linewidth=2)
axes[1, 1].set_title(r'$\operatorname{sech} x = \frac{1}{\cosh x}$', fontsize=14)
axes[1, 1].set_xlabel(r'$x$')
axes[1, 1].set_ylim(0, 1.1)

axes[1, 2].plot(x_neg, 1/np.tanh(x_neg), color='red', linewidth=2)
axes[1, 2].plot(x_pos, 1/np.tanh(x_pos), color='red', linewidth=2)
axes[1, 2].set_title(r'$\coth x = \frac{1}{\tanh x}$', fontsize=14)
axes[1, 2].set_xlabel(r'$x$')
axes[1, 2].set_ylim(-5, 5)

# Add grid, zero lines, and format axes
for i in range(2):
    for j in range(3):
        axes[i, j].axhline(y=0, color='k', linestyle='-', alpha=0.3)
        axes[i, j].axvline(x=0, color='k', linestyle='-', alpha=0.3)
        axes[i, j].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
