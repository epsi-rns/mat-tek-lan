# Integration

## Integrals

For the constant of integration C, I used the values C = -1, 0, 1 in the plots.

```python
# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(r'Integrals of Basic Trigonometric Functions', fontsize=16, fontweight='bold')

# Generate x values from -2 to 2
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Plot  sin x dx = -cos x + C
# We'll show multiple antiderivatives with different constants C
for C in [-1, 0, 1]:
    axes[0].plot(x, -np.cos(x) + C, color='blue', linewidth=2, alpha=0.7, label=f'C = {C}' if C == 0 else "")
axes[0].set_xlabel(r'$x$')
axes[0].set_ylabel(r'$-\cos x + C$')
axes[0].set_ylim(-2, 2)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0].grid(True, alpha=0.3)
axes[0].set_title(r'$\int \sin x \, dx = -\cos x + C$', fontsize=14)
axes[0].legend()

# Plot  cos x dx = sin x + C
for C in [-1, 0, 1]:
    axes[1].plot(x, np.sin(x) + C, color='teal', linewidth=2, alpha=0.7, label=f'C = {C}' if C == 0 else "")
axes[1].set_xlabel(r'$x$')
axes[1].set_ylabel(r'$\sin x + C$')
axes[1].set_ylim(-2, 2)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)
axes[1].set_title(r'$\int \cos x \, dx = \sin x + C$', fontsize=14)
axes[1].legend()

# Plot  tan x dx = -ln|cos x| + C
# Handle the absolute value and avoid division by zero
cos_x = np.cos(x)
valid_indices = (np.abs(cos_x) > 0.01)
integral_tan = np.zeros_like(x)
integral_tan[valid_indices] = -np.log(np.abs(cos_x[valid_indices]))
integral_tan[~valid_indices] = np.nan

for C in [-1, 0, 1]:
    axes[2].plot(x, integral_tan + C, color='red', linewidth=2, alpha=0.7, label=f'C = {C}' if C == 0 else "")
axes[2].set_xlabel(r'$x$')
axes[2].set_ylabel(r'$-\ln|\cos x| + C$')
axes[2].set_ylim(-5, 5)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[2].grid(True, alpha=0.3)
axes[2].set_title(r'$\int \tan x \, dx = -\ln|\cos x| + C$', fontsize=14)
axes[2].legend()

# Add vertical lines at multiples of /2 for reference
for i in range(-4, 5):
    for ax in axes:
        ax.axvline(x=i*np.pi/2, color='gray', linestyle='--', alpha=0.5)

# Add  labels to x-axis
for ax in axes:
    ax.set_xticks([-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, 
                   np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels([r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', 
                        r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])

plt.tight_layout()
plt.show()
```

```python
# Create a figure to show the area interpretation
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(r'Definite Integrals as Accumulated Area', fontsize=16, fontweight='bold')

# Generate x values
x = np.linspace(0, 2*np.pi, 1000)

# Left:  sin x dx from 0 to various points
for endpoint in [np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]:
    mask = x <= endpoint
    axes[0].fill_between(x[mask], np.sin(x[mask]), alpha=0.3, color='blue')
    area = -np.cos(endpoint) - (-np.cos(0))  #  sin x dx = -cos x
    axes[0].text(endpoint/2, 0.5, f'Area = {area:.2f}', ha='center')

axes[0].plot(x, np.sin(x), color='blue', linewidth=2, label=r'$\sin x$')
axes[0].set_title(r'$\int_0^x \sin t \, dt = 1 - \cos x$', fontsize=14)
axes[0].set_xlabel(r'$x$')
axes[0].set_ylabel(r'$\sin x$')
axes[0].set_ylim(-1.5, 1.5)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].grid(True, alpha=0.3)

# Center:  cos x dx from 0 to various points
for endpoint in [np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]:
    mask = x <= endpoint
    axes[1].fill_between(x[mask], np.cos(x[mask]), alpha=0.3, color='teal')
    area = np.sin(endpoint) - np.sin(0)  #  cos x dx = sin x
    axes[1].text(endpoint/2, 0.5, f'Area = {area:.2f}', ha='center')

axes[1].plot(x, np.cos(x), color='teal', linewidth=2, label=r'$\cos x$')
axes[1].set_title(r'$\int_0^x \cos t \, dt = \sin x$', fontsize=14)
axes[1].set_xlabel(r'$x$')
axes[1].set_ylabel(r'$\cos x$')
axes[1].set_ylim(-1.5, 1.5)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)

# Right: Show both integrals together
axes[2].plot(x, 1 - np.cos(x), color='blue', linewidth=2, label=r'$\int_0^x \sin t \, dt = 1 - \cos x$')
axes[2].plot(x, np.sin(x), color='teal', linewidth=2, label=r'$\int_0^x \cos t \, dt = \sin x$')
axes[2].set_title(r'Accumulated Area Functions', fontsize=14)
axes[2].set_xlabel(r'$x$')
axes[2].set_ylabel(r'Accumulated Area')
axes[2].set_ylim(-1.5, 2.5)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].grid(True, alpha=0.3)
axes[2].legend()

# Add  labels to x-axis
for ax in axes:
    ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])

plt.tight_layout()
plt.show()
```

## Integrals of Reciprocal

```python
import matplotlib.pyplot as plt
import numpy as np

# Use matplotlib's built-in mathtext for LaTeX-like rendering
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'serif'

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(r'Integrals of Reciprocal Trigonometric Functions', fontsize=16, fontweight='bold')

# Generate x values from -2 to 2, avoiding zeros where functions are undefined
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
x = x[(np.abs(np.sin(x)) > 0.01) & (np.abs(np.cos(x)) > 0.01)]  # Remove points near asymptotes

# Plot  csc x dx = -ln|csc x + cot x| + C
csc_x = 1 / np.sin(x)
cot_x = np.cos(x) / np.sin(x)
integral_csc = -np.log(np.abs(csc_x + cot_x))
integral_csc = np.where(np.abs(integral_csc) > 10, np.nan, integral_csc)  # Limit extreme values

for C in [-1, 0, 1]:
    axes[0].plot(x, integral_csc + C, color='blue', linewidth=2, alpha=0.7, label=f'C = {C}' if C == 0 else "")
axes[0].set_xlabel(r'$x$')
axes[0].set_ylabel(r'$-\ln|\csc x + \cot x| + C$')
axes[0].set_ylim(-5, 5)
axes[0].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[0].grid(True, alpha=0.3)
axes[0].set_title(r'$\int \csc x \, dx = -\ln|\csc x + \cot x| + C$', fontsize=14)
axes[0].legend()

# Plot  sec x dx = ln|sec x + tan x| + C
sec_x = 1 / np.cos(x)
tan_x = np.sin(x) / np.cos(x)
integral_sec = np.log(np.abs(sec_x + tan_x))
integral_sec = np.where(np.abs(integral_sec) > 10, np.nan, integral_sec)  # Limit extreme values

for C in [-1, 0, 1]:
    axes[1].plot(x, integral_sec + C, color='teal', linewidth=2, alpha=0.7, label=f'C = {C}' if C == 0 else "")
axes[1].set_xlabel(r'$x$')
axes[1].set_ylabel(r'$\ln|\sec x + \tan x| + C$')
axes[1].set_ylim(-5, 5)
axes[1].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)
axes[1].set_title(r'$\int \sec x \, dx = \ln|\sec x + \tan x| + C$', fontsize=14)
axes[1].legend()

# Plot  cot x dx = ln|sin x| + C
integral_cot = np.log(np.abs(np.sin(x)))
integral_cot = np.where(np.abs(integral_cot) > 10, np.nan, integral_cot)  # Limit extreme values

for C in [-1, 0, 1]:
    axes[2].plot(x, integral_cot + C, color='red', linewidth=2, alpha=0.7, label=f'C = {C}' if C == 0 else "")
axes[2].set_xlabel(r'$x$')
axes[2].set_ylabel(r'$\ln|\sin x| + C$')
axes[2].set_ylim(-5, 5)
axes[2].axhline(y=0, color='k', linestyle='-', alpha=0.3)
axes[2].axvline(x=0, color='k', linestyle='-', alpha=0.3)
axes[2].grid(True, alpha=0.3)
axes[2].set_title(r'$\int \cot x \, dx = \ln|\sin x| + C$', fontsize=14)
axes[2].legend()

# Add vertical lines at multiples of /2 for reference
for i in range(-4, 5):
    for ax in axes:
        ax.axvline(x=i*np.pi/2, color='gray', linestyle='--', alpha=0.5)

# Add  labels to x-axis
for ax in axes:
    ax.set_xticks([-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, 
                   np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
    ax.set_xticklabels([r'$-2\pi$', r'$-\frac{3\pi}{2}$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', 
                        r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])

plt.tight_layout()
plt.show()
```

```python
# Create a 2x3 grid of subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle(r'Comparison: Hyperbolic vs Trigonometric Integrals', fontsize=16, fontweight='bold')

# Generate x values
x_hyper = np.linspace(-3, 3, 1000)
x_trig = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Row 1: Hyperbolic integrals
axes[0, 0].plot(x_hyper, np.cosh(x_hyper), color='blue', linewidth=2, label=r'$\int \sinh x \, dx = \cosh x$')
axes[0, 0].set_title(r'Hyperbolic: $\int \sinh x \, dx = \cosh x + C$', fontsize=14)
axes[0, 0].set_ylabel(r'Integral')
axes[0, 0].set_ylim(-1, 5)

axes[0, 1].plot(x_hyper, np.sinh(x_hyper), color='teal', linewidth=2, label=r'$\int \cosh x \, dx = \sinh x$')
axes[0, 1].set_title(r'Hyperbolic: $\int \cosh x \, dx = \sinh x + C$', fontsize=14)
axes[0, 1].set_ylim(-5, 5)

integral_tanh = np.log(np.cosh(x_hyper))
axes[0, 2].plot(x_hyper, integral_tanh, color='red', linewidth=2, label=r'$\int \tanh x \, dx = \ln(\cosh x)$')
axes[0, 2].set_title(r'Hyperbolic: $\int \tanh x \, dx = \ln(\cosh x) + C$', fontsize=14)
axes[0, 2].set_ylim(-2, 3)

# Row 2: Trigonometric integrals
axes[1, 0].plot(x_trig, -np.cos(x_trig), color='blue', linewidth=2, label=r'$\int \sin x \, dx = -\cos x$')
axes[1, 0].set_title(r'Trigonometric: $\int \sin x \, dx = -\cos x + C$', fontsize=14)
axes[1, 0].set_xlabel(r'$x$')
axes[1, 0].set_ylabel(r'Integral')
axes[1, 0].set_ylim(-1.5, 1.5)

axes[1, 1].plot(x_trig, np.sin(x_trig), color='teal', linewidth=2, label=r'$\int \cos x \, dx = \sin x$')
axes[1, 1].set_title(r'Trigonometric: $\int \cos x \, dx = \sin x + C$', fontsize=14)
axes[1, 1].set_xlabel(r'$x$')
axes[1, 1].set_ylim(-1.5, 1.5)

# Handle the integral of tan x carefully
cos_x = np.cos(x_trig)
valid_indices = (np.abs(cos_x) > 0.01)
integral_tan = np.zeros_like(x_trig)
integral_tan[valid_indices] = -np.log(np.abs(cos_x[valid_indices]))
integral_tan[~valid_indices] = np.nan

axes[1, 2].plot(x_trig, integral_tan, color='red', linewidth=2, label=r'$\int \tan x \, dx = -\ln|\cos x|$')
axes[1, 2].set_title(r'Trigonometric: $\int \tan x \, dx = -\ln|\cos x| + C$', fontsize=14)
axes[1, 2].set_xlabel(r'$x$')
axes[1, 2].set_ylim(-5, 5)

# Add grid, zero lines, and format axes
for i in range(2):
    for j in range(3):
        axes[i, j].axhline(y=0, color='k', linestyle='-', alpha=0.3)
        axes[i, j].axvline(x=0, color='k', linestyle='-', alpha=0.3)
        axes[i, j].grid(True, alpha=0.3)

# Add  labels to x-axis for trigonometric functions
for j in range(3):
    axes[1, j].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
    axes[1, j].set_xticklabels([r'$-2\pi$', r'$-\pi$', r'$0$', r'$\pi$', r'$2\pi$'])

plt.tight_layout()
plt.show()
```
