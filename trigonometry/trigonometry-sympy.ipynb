## Exponential Integral

```python
# Define additional symbols
a, b = symbols('a b', real=True)

# Display as individual equations
display(Math(r'\textbf{Exponential-Trigonometric Integrals:}'))

# Define the integrals
integral1 = Integral(exp(a*x)*sin(b*x), x)
integral2 = Integral(exp(a*x)*cos(b*x), x)

# Their results
result1 = (exp(a*x)/(a**2 + b**2)) * (a*sin(b*x) - b*cos(b*x))
result2 = (exp(a*x)/(a**2 + b**2)) * (a*cos(b*x) + b*sin(b*x))

# Display the equations
display(Math(latex(integral1) + ' = ' + latex(result1) + ' + C'))
display(Math(latex(integral2) + ' = ' + latex(result2) + ' + C'))

# Verification by differentiation
display(Math(r'\textbf{Verification by Differentiation:}'))

verify1 = simplify(diff(result1, x) - exp(a*x)*sin(b*x))
verify2 = simplify(diff(result2, x) - exp(a*x)*cos(b*x))

print("Verification (should all be 0):")
print(f"First integral: {verify1}")
print(f"Second integral: {verify2}")
```
